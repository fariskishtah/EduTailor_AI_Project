```python
features = df[["studytime", "failures", "absences"]]

target = df["G3"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)

print("Model Error:", mse)
```

```python
def classify_grade(grade):
    
    if grade >= 15:
        return "Excellent"
    
    elif grade >= 10:
        return "Good"
    
    else:
        return "Needs Support"
```

```python
df["Performance"] = df["G3"].apply(classify_grade)
```

```python
df[["G3","Performance"]].head()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(y_test, predictions)

plt.xlabel("Actual Grades")
plt.ylabel("Predicted Grades")

plt.title("Actual vs Predicted Student Grades")

plt.show()
```

```python
def recommendation(level):

    if level == "Excellent":
        return "Provide advanced materials and leadership projects"

    elif level == "Good":
        return "Provide practice exercises and guided learning"

    else:
        return "Provide personalized tutoring and extra support"
```

```python
df["Recommendation"] = df["Performance"].apply(recommendation)
```
